#### Task 4.4.2: Correlation Times in the DES
What conclusions can you draw from the correlation coefficients? Interpret all results of task
4.3.1. What can you say about whether packets have to wait or not? Explain!

IAT and WT:
negative and very small to zero for all values of offered traffic. This means that they if IAT is short, packets are more likely to wait longer

IAT and Service Time:
value really close to 0 (can be either positive or negative) -> no correlation. This can be explained as both values are generated by separate RNS generators.

IAT and System time:
The values show similar behavior to the correlation value of IAT and WT, which makes sense because system time is a combination of service time and waiting time (plus service time is uncorrelated with IAT). This means that shorter IAT means higher system time.

Service time and system time
The values show higher correlation rate for low offered traffic (close to 1) and lower correlation for higher offered traffic. This is related to the previous point where system time is a combination of waiting time and service time. In the case of low offered traffic, the waiting time is not so high and therefore the value of system time is roughly equal to service time. This is not true for higher utilization because waiting time is now not close to 0.

WT1 and WT2
The waiting time of both users show a value close to 0 meaning that they are not correlated to one another for low offered traffic. However the value slowly increase as the offered traffic increase. This can be explained because if a user is waiting it means a user that will arrive will have to wait as well. This is however not apparent for low offered traffic as there is almost no waiting time or the waiting time is too short to show a correlation. (if value of wt1 increase, wt2 will also increase)
